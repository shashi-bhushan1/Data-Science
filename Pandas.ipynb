# Creating Data Set

```python
labels=['W','X','Y','Z']
sample_list=[100,200,300,400]
array=np.array([1,2,3,4])
dict={'W':10,'X':20,'Y':30,'Z':40}
```

```python
pd.Series(data=sample_list)    ### Series is used when there is one column
                               ### If there is only one element is () then is is defaultly data and indexing will be from 0
                               ### Indexing starts from zero
                               ### Capital S is there not small s
```

```python
pd.Series(data=sample_list,index=labels)         ## first elements inside () is data and second is index
```

```python
pd.Series(sample_list,array)
```

```python
pd.Series(dict)     ### In case of dict, it will automatically gets matched 
```

```python
sports1=pd.Series([1,2,3,4],index=['cricket','football','basketball','golf'])
sports1
```

```python
sports1['cricket']  
```

```python
sports2=pd.Series([11,2,3,4],index=['cricket','football','baseball','golf'])
sports2
```

```python
sports1+sports2        ### Rows with same index gets added while different will through NaN(not a number)
```

### DataFrame is used for multiple columns

```python
df1=pd.DataFrame(np.random.rand(8,5),index='A B C D E F G H'.split(),columns='Score1 Score2 Score3 Score4 Score5'.split())
df1
```

```python
# In abvoe example "A" , "B","C" etc all are separate strings and .split() will separate them 
# Also np.random.rand(8,5) will populate the data, then comes the indexing and columns= 'Score1 to Score5' will give the name of the columns
```

```python
df1["Score3"]      # To fetch a single column
```

```python
df1[['Score2','Score4','Score5']]       ## To fetch 2 or more columns we need to use 2 square braces
```

```python
df1['Score6']=df1['Score1']+df1['Score2']             ## Mutability is also possible
df1
```
