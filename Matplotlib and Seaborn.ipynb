# Matplotlib

## Visualisation

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Draw a line in a Diagram from position (0,10) to position (5,150):
xpoints=np.array([1,2])
ypoints=np.array([5,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,2])
ypoints=np.array([5,8])
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
ypoints=np.array([3,8,1,12,7,9])            ###  Here one only one axis's values is specified then these values will be plotted
                                            ##   against x axis with scale starting from 0 to N-1(total no. of elements here 6)
plt.plot(ypoints,marker='*')                ## ALso 3 with plotted against 0 and 8 with 2 and so on 9 will be against 5(N-1 or 6-1)
plt.show()
```

```python
plt.plot(ypoints,marker='X')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])

plt.plot(ypoints,'x'':r')   # : is for dotted line and r represents red color
plt.show()
```

```python
xpoints=np.array([3,8,1,10])
ypoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker='o',ms=10,color='green')        ### ms=marker size
plt.show()
```

```python
### You can use the keyword argument markeredgecolor or the shorter or the shorter mec to set the color of the edge of marker
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=20,mec='r')     ### when only ypoints is plotted, default x value is from 0 to 3 (N-1)
plt.show()
```

```python
### mfc or marker face color 
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=10,mfc='r',mec='b',color='green')
plt.show()
```

```python
### Use both the mec and mfc arguments to color of the entire marker
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='D',ms=10,mfc='k',mec='y')       ### Here marker=D shows diamond and mec=k shows black color
plt.show() 
```

```python
### linestyle=dashed shows dashed lines
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dashed',marker='o',mec='hotpink',mfc='blue',ms=20)  
plt.show()
```

```python
### linestyle=dashed shows dashed lines
plt.plot(ypoints,linestyle='dotted',marker='o',mec='red',mfc='green',ms=25)  
plt.show()
```

```python
xpoints=np.array([10,20,30])
ypoints=np.array([1,2,3])
plt.plot(xpoints,ypoints,marker='*',color='red',mfc='g',mec='r',ms=15,linestyle='dashed')
plt.show()
```

```python
### You can use the keyword argument linewidth or the shorter lw to change the width of the line.

### The value is a floating number, in points

ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linewidth='10')
plt.show()
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,5,11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

```python
### We use plt.xlabel() to label x axis and similarly y axis

x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y,color='r',linewidth=10)

plt.xlabel("Average Pulse",fontsize=10,color="red")
plt.ylabel("Calorie Burnage",fontsize=10,color="g")

plt.show()

```

```python
### With pyplot, you can use the title() function to set a title for the plot
### We use loc to set the location of title or axes labels
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])

plt.title("sports Watch Data",loc="right",fontsize=20,color="green")

plt.xlabel("Average Pulse",loc="left")
plt.ylabel("Calorie Burnage")

plt.plot(x,y)
plt.show()
```

# The subplot() function takes three arguments that describes the layout of the figure.
## The layout is organized in rows and columns, which are represented by the first and second argument.
### The third argument represents the index of the current plot.
#### plt.subplot(1,2,1) # the figure has 1 row, 2 columns, and this plot is the first plt. plt.subplot(1,2,2) # the figure has 1 row, 2 columns, and this plot is the second plot.

```python
# plot1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
# plot1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(x,y)

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

```python
## You can add a title to each plot with the title() function

# plot1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("Income")

plt.show()
```

```python
# You can add a title to the entire figure with the suptitle() function

# plot1:
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("SALES")

#plot2:
x=np.array([0,1,2,3])
y=np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

## We use plt.scatter() function to make scatter plot

```python
## With Pyplot, you can use the scatter() function to draw a scatter plot.

## The scatter() function plots one dot for each observation. It needs two arrays of the same length,
## one for the values of the x-axis, and one for values on the y-axis
x=np.array([1,2,3,4,5])
y=np.array([5,10,15,20,25]) 
plt.scatter(x,y)
```

```python
x=np.array([1,2,3,4,5])
y=np.array([25,20,15,10,5])
plt.scatter(x,y)
```

```python
# To draw multiple scatter plot in a single plot, we'll have to use plt.scatter(x,y) function individually and then plt.show()

#day one, the age and speed of 13 cars:
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,111,86,103,87,94,78,77,85,86,12])
plt.scatter(x,y)

#day two, the age and speed of 15 cars:
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)

plt.show()
```

```python
#day one, the age and speed of 13 cars:
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,111,86,103,87,94,78,77,85,86,12])
plt.scatter(x,y,color='green')

#day two, the age and speed of 15 cars:
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y,color='hotpink')

plt.show()
```

```python
# We can change the color of each data point by passing the argument either color or c
# But we can change the size of the the data point by passing s argument instead of size 
x=np.array([1,2,3,4,5])
y=np.array([6,7,8,9,10])
colors=(['red','yellow','black','blue','green'])
sizes=([10,200,30,40,50])
plt.scatter(x,y,c=colors,s=sizes)
```

```python
# ColorMap
## The Matplotlib module has a number of available colormaps.
### A colormap is like a list of colors, where each color has a value that ranges from 0 to 100.

### Here is an example of a colormap:

x=np.array([5,7,8,9,33,35])
y=np.array([88,99,37,94,92,100])
colors=np.array([0,10,20,30,40,55])

plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
## here in np.random.normal(), the firsr argument is for mean, second is for standard distribution and third is for size
## mean will tell us about the loc of the center and standard distribtution will tell about the spread of the data

## To draw a histogram use argument plt.hist()

x=np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
## Use plt.pie() to draw a pie chart

y=np.array([10,135,25,15,19])

plt.pie(y)
plt.show()
```

```python
## Add labels to the pie chart with the label parameter.

### The label parameter can be an array or a list with one label for each wedge
### use argument labels and not label
y=np.array([18,25,15,10])
mylabels=["Cherries","Apples","Bananas","Dates"]

plt.pie(y,labels=mylabels)
plt.show()
```

```python
### As mentioned, the default start angle is at the x-axis, but you can change the start angle 
### by specifying a startangle parameter.

## The startangle parameter is defined with an angle in degrees, default angle is 0:

y=np.array([35,25,25,15])
mylabels=np.array(["Apples","Bananas","Cherries","Dates"])

plt.pie(y,labels=mylabels,startangle=90)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=np.array(["Apples","Bananas","Cherries","Dates"])
myexplode=([0,0.95,0.1,3.2])

plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels=np.array(["Apples","Bananas","Cherries","Dates"])
mycolors=["black","hotpink","b","yellow"]

plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

```python
## To show legends , use plt.legends() function

y=np.array([35,25,25,15])
mylabels=np.array(["Apples","Bananas","Cherries","Dates"])


plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```
